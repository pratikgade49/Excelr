```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
mv=pd.read_csv('../input/movies/my_movies.csv')
```

```python
mv
```

```python
mv.shape
```

```python
#Dropping unnecessary Columns
mv=mv.drop(['V1','V2','V3','V4','V5'], axis=1)
mv
```

```python
#Apriory Algorithm
#with min_support of 0.2
frequent_itemsets1=apriori(mv,min_support=0.2,use_colnames=True)
frequent_itemsets1
```

```python
rules1= association_rules(frequent_itemsets1, metric="lift", min_threshold=0.7)
rules1
```

```python
rules1.sort_values('lift',ascending = False)[0:20]
```

```python
rules1.sort_values('lift',ascending = False)[0:20]
```

```python
rules1[rules1.lift>1]
```

```python
df=pd.DataFrame(data=frequent_itemsets1)
df
df.duplicated()
```

```python
# visualization of obtained rule
rules1.plot(kind='bar',x='support',y='confidence',color='grey')
plt.title('Borplot')
plt.xlabel('support')
plt.ylabel('confidence')
plt.figure(figsize=(30,10))
```

```python
# visualization of obtained rule
plt.scatter(rules1.support,rules1.confidence)
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
